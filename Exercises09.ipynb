## 1. Curve Fitting: The $M_\mathrm{BH}-\sigma_*$ Relation
**`Carroll and Ostlie (2017). Section 25.2`**

In 2006, Greene and Ho studied the characteristics of 88 galaxies to show that there is an apparent relationship between the stellar velocity dispersion $\sigma_*$ in a galaxy bulge and the mass $M_\mathrm{BH}$ of the supermassive black hole at its
center, as shown in the Figure below. This  is known as the $M_\mathrm{BH}-\sigma_*$ relation and is not yet  completely understood. 


Greene, J. E. and Ho, L. C. *The MBH − σ∗ Relation in Local Active Galaxies*. ApJ 641 L21 (2006)
https://ui.adsabs.harvard.edu/abs/2006ApJ...641L..21G/abstract

The dataset used by Greene and Ho is available online in various formats at

http://vizier.cfa.harvard.edu/viz-bin/VizieR?-source=J/ApJ/641/L21.

Following Greene and Ho, we will make a linear fit between the variables $\log_{10} M$ and $\log_{10} \left( \frac{\sigma_*}{\sigma_0} \right)$,  where the constant $\sigma_0 = 200$ is a reference value given by the authors. 

1. Add a column for the $\log_{10} \left( \frac{\sigma_*}{\sigma_0} \right)$ and make a scatter plot showing the data $\log_\mathrm{10} M$ as a function of $\log_\mathrm{10}
\left( \frac{\sigma_*}{\sigma_0} \right)$.

2. Compute the covariance and correlation coefficients to study the behavior of these features. 

3. Implement a linear regression fit to the data (ignoring errors). Make a plot of data and fit. Compare your fit results to that of Greene \& Ho (2006).

3. Include the errors in $\log_\mathrm{10} \sigma_*$ and $\log_\mathrm{10} M_\mathrm{BH}$ in your fit (make sure that the errors are defined properly!). There may be multiple errors given, use your judgement which ones to pick. How does this change your fit?
